## Моделирование

### Работа с признаками

Первым напрашивается длительность обслуживания через признаки EndDate и BeginDate, однако мы используем EndDate для формирования таргета, кроме того даты конца обслуживания для неушедших клиентов у нас нет, а пропуски на том месте мы запонили датой выгрузки. Все в месте обеспечит нам гарантированную утечку, так делать нельзя.

Удалим столбцы customerID, BeginDate, EndDate.

```python
full_data = full_data.drop(['customerID', 'BeginDate', 'EndDate'], axis=1)
```

Перейдем к анализу взаимосвязи признаков, учтем, что не все они являются количественными, то есть обычные коэффициент корреляции Пирсона будет работать некорректно, воспользуемся библиотекой phik.

```python
data_types = {'Type': 'categorical',
              'PaperlessBilling':'categorical',
              'Partner':'categorical', 
              'Dependents':'categorical', 
              'InternetService':'categorical', 
              'OnlineSecurity':'categorical', 
              'OnlineBackup':'categorical', 
              'DeviceProtection':'categorical', 
              'TechSupport':'categorical', 
              'StreamingTV':'categorical', 
              'StreamingMovies':'categorical', 
              'DeviceProtection':'categorical', 
              'MultipleLines':'categorical',
              'MonthlyCharges':'interval',
              'TotalCharges':'interval',
              } 
interval_cols = [col for col, v in data_types.items() if v=='interval' and col in full_data.columns] 
interval_cols 
```

```python
phik_overview = full_data.phik_matrix(interval_cols=interval_cols) 
sns.heatmap(phik_overview) 
phik_overview
```

Заметно, что признаков с сильной связью с целевым признаком у нас в принципе нет, при это есть целый квадрат мультиколлинеарности. Оставим только один из коллинеарных признаков по принцепу самой выской связи с таргетом.

```python
def drop_multicollinear_by_phik(
    df: pd.DataFrame,
    target: str,
    threshold: float = 0.7,
    interval_cols: list = None,
):
    
    phik_corr = df.phik_matrix(
        interval_cols=interval_cols,
    )
    
    
    features = [c for c in df.columns if c != target]
    
    
    corr_with_target = (
        phik_corr[target]
        .abs()
        .drop(index=target)
        .loc[features]
        .sort_values(ascending=False)
    )
    
    selected = []
    for feat in corr_with_target.index:
       
        conflicts = [
            phik_corr.loc[feat, sel] 
            for sel in selected
        ]
        if not conflicts or max(np.abs(conflicts)) < threshold:
            selected.append(feat)
    
    dropped = [f for f in features if f not in selected]
    return selected, dropped
```

```python
selected, dropped = drop_multicollinear_by_phik(full_data, 'label', 0.7, interval_cols)
selected
```

Окей, это признаки с которыми мы будет работать.

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(full_data[selected + ['label']].phik_matrix(interval_cols=interval_cols) , annot=True);
```

Уже неплохо. Следует уточнить, что деревянные модели, которые мы собираемся использовать, сами неплохо справляются с отбором признаков, а вот через зависимую от таргета фильтрацию могло что-то утечь, поэтому первоначально мы подадим в модели все фичи, если будут проблемы, вернемся к их отбору внутри pipeline.

### Деление на выборки

```python
features_train, features_test, target_train, target_test = train_test_split(
                                                                            full_data.drop('label', axis=1), full_data['label'], 
                                                                            test_size=0.25, random_state=170723,
                                                                            stratify=full_data['label']
                                                                            ) 
```

### Кодирование признаков

Будем использовать две модели, CatBoost и случайный лес. Для CatBoost создадим параметр cat_features и pool, последний сильно ускоряет обучение. Почти наверняка бустинг будет лучше, но интересно посмотреть насколько.

```python
cat_features = ['PaymentMethod', 'Partner', 'MultipleLines', 'Type', 'OnlineBackup', 
                'SeniorCitizen', 'PaperlessBilling', 'Dependents', 'gender', 'InternetService', 
                'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'] 
```

```python
train_pool=Pool(features_train, label=target_train, cat_features=cat_features)
```

Категориальные признаки для леса создадим про помощи OE, для него напишем пайплайн. Вообще, с OE вряд ли что-то утечет на кросс-валидации, но все-таки.

```python
random_forest_model = Pipeline([
                                ("transformer", ColumnTransformer([("encoder", OrdinalEncoder(), cat_features)])),
                                ("classifier", RandomForestClassifier(random_state=170723)) 
                              ])
```

### Обучение моделей

#### CatBoost

```python
CatBoost_model = CatBoostClassifier(loss_function='Logloss', 
                                    eval_metric='AUC',
                                    class_weights=[1.5, 1], # сделано намеренно
                                    verbose=False, 
                                    random_seed=170723) 
params = {'depth': [2, 4, 8],
          'learning_rate' : [0.15, 0.1, 0.03], 
          'l2_leaf_reg': [2, 4, 8],
          'iterations': [500],
          'one_hot_max_size': [10, 15],
          } 


CatBoost_grid = CatBoost_model.grid_search(params, train_pool, cv=5, partition_random_seed=170723) 
```

```python
CatBoost_grid['params']
```

```python
max(CatBoost_grid['cv_results']['test-AUC-mean'])
```

Бустинг показал результат порядка 0.81, результат хороший. 

#### Случайный лес

```python
params_RF = {
             'classifier__n_estimators': [40, 80, 100],
             'classifier__max_depth': [2, 4, 8],
             'classifier__max_features': [2, 4], 
             'transformer__remainder': ['passthrough']
             }
```

```python
grid_RF = GridSearchCV(random_forest_model, param_grid = params_RF, scoring='roc_auc', cv=5)
```

```python
grid_RF.fit(features_train, target_train)
```

```python
grid_RF.best_score_
```

Случайный лес отработал с результатом 0.79, что довольно тоже неплохо, несколько ниже бустинга.

#### Вывод по обучению моделей

Таким образом, лучшей моделью ожидаемо становится CatBoost с результатом 0.81 по AUC-ROC на кросс-валидации. 

### Тестирование лучшей модели

```python
roc_auc_score(target_test, CatBoost_model.predict_proba(features_test)[:,1])
```

Итого, CatBoost на тестовой выборке показывает результат порядка 0.82 по метрике AUC-ROC.

Как можно заметить, с далеким отрывом самыми важными признакоми в данной модели являются TotalCharges и MonthlyCharges, иначе говоря, если клиент уже давно платит нам деньги и пользуется множеством услуг, вряд ли он куда-то уйдет.

```python
best_thres = 0.01
max_f1_score = 0
for i in range(1, 100):
    thres = i / 100
    var = f1_score(target_train, list(map(int, probs[:, 1] >= thres)))
    if var > max_f1_score:
        max_f1_score = var
        best_thres = thres
```

```python
print(best_thres) 
display(max_f1_score)
```

```python
print('Recall:', recall_score(target_train, list(map(int, probs[:, 1] >=  best_thres))))
print('Precision:', precision_score(target_train, list(map(int, probs[:, 1] >=  best_thres))))
print('AP:', average_precision_score(target_train, probs[:, 1]))
```

Таким образом, с при веротяности отсечки в 0.23, модель имеет результат по f1 - 0.47, который состоит из 0.56 Recall и 0.4 Precision, примерно 0.47 по average_precision. При имеющихся данных - это неплохо.

```python
CatBoost_model.set_probability_threshold(best_thres) 
f1 = f1_score(target_test, CatBoost_model.predict(features_test))
auc_roc = roc_auc_score(target_test, CatBoost_model.predict_proba(features_test)[:,1])
auc_pr = average_precision_score(target_test, CatBoost_model.predict_proba(features_test)[:,1])

print('AUC-ROC:', auc_roc)
print('F1:', f1)
print('AP:', auc_pr)
```

На тесте значение ROC_AUC аналогичное - 0.83, f1 чуть выше - 0.48, average_precision порядка 0.51, сильно выше, чем на трейне.
